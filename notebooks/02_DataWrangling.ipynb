```python
import pandas as pd

raw_dec_benign = pd.read_csv('../data/raw/dec/decimal_benign.csv')
raw_dec_dos = pd.read_csv('../data/raw/dec/decimal_DoS.csv')
raw_dec_gas = pd.read_csv('../data/raw/dec/decimal_spoofing-GAS.csv')
raw_dec_rpm = pd.read_csv('../data/raw/dec/decimal_spoofing-RPM.csv')
raw_dec_speed = pd.read_csv('../data/raw/dec/decimal_spoofing-SPEED.csv')
raw_dec_wheel = pd.read_csv('../data/raw/dec/decimal_spoofing-STEERING_WHEEL.csv')
```

```python
# Concatenar verticalmente  
df = pd.concat([  
    raw_dec_benign,  
    raw_dec_dos,  
    raw_dec_gas,  
    raw_dec_rpm,  
    raw_dec_speed,  
    raw_dec_wheel  
], axis=0)  
# Verificar estructura inicial  
print("Dimensiones del dataset combinado:", df.shape)  
df.head()  

#Se combinan datasets separados (benignos y ataques) para crear un único DataFrame.
#Por qué: Facilitar el procesamiento uniforme y evitar sesgos en futuros pasos.
```

```python
df['ID_freq'] = df['ID'].map(df['ID'].value_counts())
df.head()  
#ID_freq: Los ataques generan alta frecuencia en un ID específico (ej. 291).
```

```python
# Calcular duplicados
duplicados = df.duplicated().sum()
df = df.drop_duplicates()
print(f"Duplicados eliminados: {duplicados}")

#Por qué eliminar duplicados:
#En el EDA, el 99.6% de las filas benignas estaban duplicadas, lo que podría inflar métricas y sesgar modelos.
#En ataques, los duplicados reflejan patrones reales (ej. DoS repite el mismo mensaje). Sin embargo, se eliminan para evitar sobreajuste.
```

```python
# Feature: Frecuencia
print("Dimensiones del dataset reducido:", df.shape)  
df.head()  
```

```python
df.to_csv('../data/clean/df_feat_engineering.csv', index=False)
```
